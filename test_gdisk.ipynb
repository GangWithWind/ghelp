```python
import numpy as np

a = np.linspace(40, 80, 100)
inc = 0.2 - ((a - 80) / 80) ** 2 * 0.1
omega = np.sin((a-40)/40 * np.pi * 2) * 10 / 180 * np.pi
import matplotlib.pyplot as plt

p = inc * np.sin(omega)
q = inc * np.cos(omega)

inc_profile = np.stack([a, p, q]).T

np.savetxt('inc_profile.txt', inc_profile)
```
